```python
class SVM_Classifier:

    # initiating the hyperparameters
    def __init__(self, learning_rate, no_of_iterations, lambda_param):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations
        self.lambda_param = lambda_param

    # Fitting the data to SVM Classifier
    def fit(self, X, Y):
        # m represents total number of datapoints in the dataset (number of rows)
        # n represents the number of input features in the dataset (number of columns)
        # We dont count outcome column in n
        self.m, self.n = X.shape

        # Initializing the weights and bias value
        # Random values as 0
        # Creating numpy array with number of columns as n
        self.w = np.zeros(self.n)
        # weight is generally array and bias is generally a scalar
        self.b = 0

        self.X = X

        self.Y = Y

        # Building Gradient Descent
        # The difference between true value and predicted value for one single data point is called loss
        # If you do that for the entire dataset that is called as cost function

        # Implementing Gradient Descent algorithm for Optimization
        for i in range(self.no_of_iterations):
            self.update_weights()

    # Function for updating the weights and bias value
    def update_weights(self):
        # Label encoding. If 0 changing into -1
        y_label = np.where(self.Y <= 0, -1, 1)

        # Gradient (dw, db) for each sample
        for index, x_i in enumerate(self.X):
            # The decision function is y = w.x + b

            condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

            if condition == True:
                dw = 2 * self.lambda_param * self.w
                db = 0

            else:
                dw = 2 * self.lambda_param * self.w - np.dot(x_i, y_label[index])
                db = y_label[index]

            # Updating the weights and bias value
            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db

    #  Function for predicting the output
    def predict(self, X):
        output = np.dot(X, self.w) - self.b
        predicted_labels = np.sign(output)
        y_hat = np.where(predicted_labels <= -1, 0, 1)
        return y_hat
```

#### Importing the libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#### Data Collection and Processing

```python
# Loading the data from csv file to pandas dataframe
diabetes_data = pd.read_csv('diabetes.csv')
diabetes_data.head()
```

```python
# Number of rows and columns in the dataset
diabetes_data.shape
```

```python
# Get the statistic measures of the dataset
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

```python
# Separating the features and target
features = diabetes_data.drop(columns='Outcome', axis=1)
target = diabetes_data['Outcome']
```

```python
print(features)
```

```python
print(target)
```

```python
# Data Standardization
scaler = StandardScaler()
scaler.fit(features)
```

```python
standarized_data = scaler.transform(features)
```

```python
print(standarized_data)
```

```python
features = standarized_data
target = diabetes_data['Outcome']
```

#### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X_test.shape)
```

#### Training the Model

```python
classifier = SVM_Classifier(learning_rate=0.0001, no_of_iterations=1000, lambda_param=0.000001)
```

```python
classifier.fit(X_train, Y_train)
```

#### Model Evaluation

```python
# Accuracy on Training Data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

print(training_data_accuracy)
```

```python
# Accuracy on Test Data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print(test_data_accuracy)
```
